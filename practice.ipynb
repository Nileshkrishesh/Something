```python
df_cleaned=pd.read_csv("StudentsPerformance.csv") 
```

```python
df_cleaned.head()
```

```python
df_cleaned.dropna(inplace=True) # drop the null values from the data
```

```python
df_cleaned.shape
```

```python
df_cleaned.drop_duplicates(inplace=True) # drop the duplicate values from the data
```

```python
df_cleaned.dtypes #dtypes is used to check the data types of the columns
```

```python
df_cleaned = pd.get_dummies(df_cleaned, drop_first=True) # get_dummies is used to convert the categorical columns into numerical columns and drop_first is used to drop the first column of the categorical columns
```

```python
df_cleaned.head()
```

```python
df_cleaned = df_cleaned.astype(int)
```

```python
df_cleaned
```

```python
from sklearn.preprocessing import StandardScaler 
from scipy.stats import pearsonr # StandardScaler is used to scale the data 
cols=['reading score','writing score'] # list of columns to be scaled
scaler = StandardScaler() # create an object of StandardScaler
df_cleaned[cols] = scaler.fit_transform(df_cleaned[cols]) # fit_transform is used to fit the data and transform it
```

```python
df_cleaned
```

```python
df_cleaned.columns
```

```python
selected_features = [ 'reading score', 'writing score', 'gender_male',
       'race/ethnicity_group B', 'race/ethnicity_group C',
       'race/ethnicity_group D', 'race/ethnicity_group E',
       "parental level of education_bachelor's degree",
       'parental level of education_high school',
       "parental level of education_master's degree",
       'parental level of education_some college',
       'parental level of education_some high school', 'lunch_standard',
       'test preparation course_none']

correlations = {
    feature: pearsonr(df_cleaned[feature], df_cleaned['math score'])[0]
    for feature in selected_features
}
correlation_df=pd.DataFrame(list(correlations.items()),columns=['Feature','Pearson Correlation'])
correlation_df.sort_values(by='Pearson Correlation',ascending=False)
# The features with the highest absolute correlation values are likely to be the most important features for predicting the target variable.    
# >0.5 or <-0.5 are considered strong correlation, 0.3 to 0.5 or -0.3 to -0.5 are considered moderate correlation, and 0.1 to 0.3 or -0.1 to -0.3 are considered weak correlation.
```

```python
cat_features =['reading score', 'race/ethnicity_group D','writing score', 'lunch_standard', 'test preparation course_none','parental level of education_high school']
```

```python
from scipy.stats import chi2_contingency # chi2_contingency is used to perform the chi-square test of independence between two categorical variables
import pandas as pd
alpha= 0.05 # alpha is the significance level for the chi-square test
df_cleaned['Math score_bin']= pd.qcut(df_cleaned['math score'], q=4, labels=False) # qcut is used to create categories based on the quantiles of the column
chi2_results={}

for col in cat_features:
    contingency_table = pd.crosstab(df_cleaned[col], df_cleaned['Math score_bin']) # crosstab is used to create a contingency table for the chi-square test
    chi2, p, dof, expected = chi2_contingency(contingency_table) # chi2_contingency is used to perform the chi-square test and returns the chi-square statistic, p-value, degrees of freedom, and expected frequencies
    decion ='Reject Null (keep feature)' if p<alpha else 'Accept Null (Drop Feature)'
    chi2_results[col] = {
        'chi2_statistic': chi2,
        'p_value': p,
        'Decision': decion
    }

chi2_df = pd.DataFrame.from_dict(chi2_results, orient='index') # from_dict is used to create a dataframe from a dictionary
chi2_df.sort_values(by='p_value') # sort_values is used to sort the values
chi2_df
```

```python
final_df = df_cleaned[[
    'reading score', 'writing score', 'lunch_standard', 'test preparation course_none','parental level of education_high school'
]]
```

```python
final_df.head()
```

```python
from sklearn.model_selection import train_test_split # train_test_split is used to split the data into training and testing sets
```

```python
X = final_df
y = df.loc[final_df.index, 'math score'] # loc is used to select the rows and columns from the dataframe
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=42) # train_test_split is used to split the data into training and testing sets, test_size is the proportion of the data to be used for testing, and random_state is used to ensure that the split is reproducible. 
```

```python
from sklearn.linear_model import LinearRegression # linear_model is used to import the linear regression model
```

```python
model=LinearRegression() # create an object of LinearRegression
model.fit(X_train,y_train) # fit is used to train the model on the training data
```

```python
y_pred=model.predict(X_test) # predict is used to make predictions on the testing data
```

```python
from sklearn.metrics import r2_score # metrics is used to import the  r2_score functions to evaluate the performance of the model 
r2=r2_score(y_test,y_pred)
n= X_test.shape[0] # shape is used to get the number of rows and columns in the dataframe, [0] is used to get the number of rows
p=X_test.shape[1] # shape is used to get the number of rows and columns in the dataframe, [1] is used to get the number of columns
adjusted_r2=1-(1-r2)*(n-1)/(n-p-1) # adjusted_r2 is used to calculate the adjusted r2 score
adjusted_r2
```
